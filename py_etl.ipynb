#### Cleaning and treating VCS data

```python
import pandas as pd
import re
from datetime import datetime
```

#### Retirements

```python
retired_df = pd.read_excel('vcus_retired.xlsx')
retired_df.head()
```

```python
retired_columns = retired_df.columns.tolist()

retired_columns
```

```python
dropped_columns = [
    'Issuance Date',
    'Sustainable Development Goals',
    'Vintage Start',
    'Methodology',
    'Total Vintage Quantity',
    'Serial Number',
    'Additional Certifications',
    'Retirement Reason',
    'Retirement Details'
]
```

```python
retired_df.drop(columns=dropped_columns, inplace=True)

retired_df
```

```python
retired_df.info()
```

```python
def replace_chars_in_column(df, column_name, replacement_char=' '):
    non_normal_pattern = re.compile('[^\x00-\x7F]')  # Matches any character outside the ASCII range (non-ASCII characters)
    
    # Apply the function to the specified column in the DataFrame
    retired_df['Retirement Beneficiary'] = retired_df['Retirement Beneficiary'].apply(lambda x: non_normal_pattern.sub(replacement_char, str(x)))
    
    return df


retired_df = replace_chars_in_column(retired_df, retired_df['Retirement Beneficiary'])
```

```python
retired_df.info()
```

```python
retired_df.to_csv(r'C:\Users\Victor\Documents\Programming\Portfolio\Power BI\DB\retired_df.csv', index=False, encoding='utf-8')
```
